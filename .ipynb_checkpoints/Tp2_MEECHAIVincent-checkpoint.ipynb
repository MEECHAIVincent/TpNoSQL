```python
from pprint import pprint
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.gym
gymnases = db.Gymnases
sportifs = db.Sportifs
print('connexion success')
```

```python
#1) Afficher toutes les collections de la base, Compter le nombre de documents ds chacune des collections, Afficher un document de chacune des collections ?


print("List of collections\n--------------------")
#list the collections
for coll in db.list_collection_names():
    print(coll)
    
print("\nNombre de documents\n--------------------")
#nombre de doc.
request = gymnases.find()
Qry = sportifs.find()
print("Sportifs : ", Qry.count())
print("Gymnases : ", request.count())

print("\nAffichage d'un document\n--------------------")
#nombre de doc.
request = gymnases.find_one()
Qry = sportifs.find_one()
print("\nDoc d'un Sportif: \n--------------------")
pprint( Qry)
print("\nDoc d'un Gym: \n--------------------")
pprint(request)
```

```python
# 2) On cherche le (ou les) conseiller(s) s’appelant "KERVADEC" ?

print("Search name 'KERVADEC'\n--------------------")
qry = sportifs.find_one({"Nom":"KERVADEC"})
pprint(qry)
```

```python
# 3) On cherche les sportifs qui ont ou bien 32 ans ou 40 ans ?

print("Search sportifs age 32 or 40\n--------------------")
qry = sportifs.find({"$or" : [{"Age": 32}, {"Age": 40}]})
for sportif in qry:
    pprint(sportif)
```

```python
# 4) On cherche les sportifs jouant au Basket-ball ?

print("Search sportifs play Basket-ball\n--------------------")
qry = sportifs.find({ "Sports.Jouer": {"$eq" : "Basket ball"}})
for sportif in qry:
    pprint(sportif)
```

```python
# 5) On cherche les sportifs ayant au moins 32 ans, ou de sexe féminin ?

print("Search sportifs min 32 or sexe female\n--------------------")
qry = sportifs.find({"$or" : [{"Age": {"$gte" : 32}}, {"Sexe": "F"}]})
for sportif in qry:
    pprint(sportif)
```

```python
# 6) Trouver le nombre de sportifs de sexe féminin dans la base ?

print("Search nombre sportifs sexe female\n--------------------")
qry = sportifs.find({"Sexe": "F"})
print(qry.count())
```

```python
# 7)	Quels sont les sportifs (identifiant, nom et prénom) qui ont entre 20 et 30 ans ?

print("Search sportifs min 20 and max 30\n--------------------")
qry = sportifs.find({ "Age": {"$gte": 20, "$lte": 30} },{"Nom":1,"Prenom":1})
for sportif in qry:
    pprint(sportif)
```

```python
# 8)	Quels sont les sportifs (identifiant et nom) qui pratiquent du hand ball ?

print("Search sportifs play Hand ball\n--------------------")
qry = sportifs.find({ "Sports.Jouer": {"$eq" : "Hand ball"}},{"Nom":1})
for sportif in qry:
    pprint(sportif)
```

```python
# 9)	Quels sont les gymnases de “Villetaneuse” ou de “Sarcelles” qui ont une surface de plus de 400 m2 ?

print("Search gym Villetaneuse or Sarcelles surface + 400m²\n--------------------")
qry = gymnases.find({"$or" : [{"Ville": "VILLETANEUSE" },{"Ville": "SARCELLES"}], "Surface": {"$gt" : 400}})
for gymnase in qry:
    pprint(gymnase)
```

```python
# 10)	Dans quels gymnases peut-on jouer au hockey le mercredi après 15H ?

print("Search gym play hockey mercredi after 15h\n--------------------")
qry = gymnases.find({"Seances.Libelle": {"$eq":"Hockey"},"Seances.Jour": {"$eq":"mercredi"},"Seances.Horaire":{"$gt":15.0}})
for gymnase in qry:
    pprint(gymnase)
```

```python
# 11)	Quels sportifs (identifiant et nom) ne pratiquent aucun sport ?

print("Search sportifs don't play sport\n--------------------")
qry = sportifs.find({ "Sports.Jouer": {"$exists" : False}},{"Nom":1})
for sportif in qry:
    pprint(sportif)
```

```python
# 12)	Quels gymnases n’ont pas de séances le dimanche ?

print("Search gym no seance dimanche\n--------------------")
qry = gymnases.find({ "Seance.Jour": { "$nin": ["dimanche", "Dimanche"] } })
for gymnase in qry:
    pprint(gymnase)
```

```python
# 13)	Quels gymnases ne proposent que des séances de basket ball ou de volley ball ?

print("Search gym no seance basket ball or volley ball\n--------------------")
qry = gymnases.find({ "Seances.Libelle": { "$all": ["Basket ball", "Volley ball"] } })
for gymnase in qry:
    pprint(gymnase)
```

```python
# 14)	Quels sont les entraîneurs qui sont aussi joueurs ?

print("Search trainer also player\n--------------------")
qry = sportifs.find({"Sports.Entrainer":{"$exists":True}})
print(qry.count())
for sportif in qry:
    pprint(sportif)
```

```python
# 15)	Quels sont les sportifs qui sont des conseillers ?

arrayConseiller = []
conseillers = sportifs.find({"IdSportifConseiller": {"$exists": True}}, {"IdSportifConseiller": 1, "_id": 0}).sort("IdSportifConseiller", 1)
for conseiller in conseillers:
    if conseiller not in arrayConseiller:
        arrayConseiller.append(conseiller)
        request = sportifs.find_one({"IdSportif": conseiller["IdSportifConseiller"]})
        pprint(request)
```

```python
# 16)	Pour le sportif “Kervadec” quel est le nom de son conseiller ?

print("Search sportif Kervadec's conseiller\n--------------------")
conseiller = sportifs.find_one({"Nom":"KERVADEC"},{"_id":0})
print(conseiller,"\n")
qry = sportifs.find_one({"IdSportif": conseiller['IdSportifConseiller']},{"_id":0})
print(qry)
```

```python
# 17)	Quels entraîneurs entraînent du hand ball et du basket ball ?

print("Search trainer also conseillers\n--------------------")
qry = sportifs.find({"Sports.Entrainer": { "$all": ["Hand ball", "Volley ball"] }})
for sportif in qry:
    pprint(sportif)
```

```python
# 18)	Quels sportifs n’ont pas de conseillers ?

print("Search trainer also conseillers\n--------------------")
qry = sportifs.find({"IdSportifConseiller":{"$exists":False}})
for sportif in qry:
    pprint(sportif)
```
